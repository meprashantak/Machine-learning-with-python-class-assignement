```python
# Assignement 1
1)What is Python and why is it called an interpreted language?
* Python is a high-level, general-purpose programming language used for web development, data science, 
  automation, artificial intelligence, and many other applications.
*It is called an interpreted language because the code is executed line by line by an interpreter
 instead of being compiled completely before running.

2)What are the key features of Python that make it popular?
*Easy to learn and read 
• Interpreted language 
• Cross-platform – runs on Windows, Linux, Mac
• Large standard library 
• Open source 
• Supports multiple programming styles 
• Strong community support

3)What is the difference between Python 2 and Python 3?
Feature	             Python 2	             Python 3
Print statement  	 print "Hello"	         print("Hello")
Unicode support	     Limited	             Full Unicode support
Division	         5/2 = 2	             5/2 = 2.5
Development	         No longer supported	 Actively developed

4)What are Python’s applications in real-world projects?
Python is used in many real-world fields:
1. Web Development - Frameworks: Django, Flask
2. Data Science & Data Analysis - Libraries: Pandas, NumPy, Matplotlib
3. Artificial Intelligence & Machine Learning - Libraries: TensorFlow, Scikit-learn
4. Automation / Scripting  - Automating repetitive tasks
5. Game Development  - Libraries like Pygame
6. Cybersecurity & Networking

5)What is PEP 8 and why is it important?
* PEP 8 stands for Python Enhancement Proposal 8.
* It is the official style guide for writing Python code.
  It defines rules like:
• Proper indentation (4 spaces)
• Naming conventions (snake_case for variables)
• Line length (max 79 characters)
• Code readability guidelines

6)Who developed Python and in which year was it released?
* Python was created by Guido van Rossum.
* It was first released in 1991.

7)What do you mean by “dynamically typed” in Python?
* In Python, you do not need to declare the variable type.
  The type is automatically determined at runtime.
Example:
x = 10      # integer
x = "Hello" # string
The same variable can store different types of values.
This is called dynamic typing.

8)What is the difference between a compiler and an interpreter?
       Compiler	                              Interpreter
 Converts whole program at once	      Converts code line by line
 Faster execution	                  Slower execution
 Errors shown after compilation	      Errors shown immediately
 Example: C, C++	                  Example: Python
```

```python
# Assigement 2
#Write a Python program that takes the length and width of a rectangle from the user and prints its area.
length = float(input("Enter the length of rectangle: "))
width = float(input("Enter the width of rectangle: "))

area = length * width

print("Area of rectangle is:", area)
```

```python
#Write a program that asks the user for the side of a square and prints its perimeter.
side = float(input("Enter the side of square: "))

perimeter = 4 * side

print("Perimeter of square is:", perimeter)
```

```python
#Take the base and height of a triangle as input and print its area
base = float(input("Enter the base of triangle: "))
height = float(input("Enter the height of triangle: "))
area = (base * height) / 2
print("Area of triangle is:", area)
```

```python
#Write a program that asks the user for the radius of a circle and prints its circumference. (Use 3.14 for π).
radius = float(input("Enter the radius of circle: "))
circumference = 2 * 3.14 * radius
print("Circumference of circle is:", circumference)
```

```python
#Take Principal (P), Rate (R), and Time (T) as input from the user and print the Simple Interest
P = float(input("Enter the Principal amount: "))
R = float(input("Enter the Rate of interest: "))
T = float(input("Enter the Time: "))

SI = (P * R * T) / 100

print("Simple Interest is:", SI)
```
